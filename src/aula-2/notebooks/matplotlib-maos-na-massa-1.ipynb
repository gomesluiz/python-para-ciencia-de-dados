11. Mude o tamanho das "círculos" do gráfico scatter com base na população armazenada no dataframe **populacoes**.

```python

titulo="Desenvolvimento Humano em 2017"
eixox="Expectiva de vida em anos"
eixoy="Renda Percapita em US$"
plt.xscale('log')
plt.xlabel(eixox)
plt.ylabel(eixoy)
plt.title(titulo)
valores    = [1000, 10000, 100000]
labels     = ["1k", "10k", "100k"]
plt.xticks(valores, labels)
np_pops = np.array(populacoes['valores'])
np_pops = np_pops * 2 
plt.scatter(percapitas['valores'], expectativas['valores'], s=np_pops)
```

12. Mude as cores das "círculos" do gráfico scatter com base no dataframe **cores**.

```python
import numpy as np
titulo="Desenvolvimento Humano em 2017"
eixox="Expectiva de vida em anos"
eixoy="Renda Percapita em US$"
plt.xscale('log')
plt.xlabel(eixox)
plt.ylabel(eixoy)
plt.title(titulo)
valores    = [1000, 10000, 100000]
labels     = ["1k", "10k", "100k"]
plt.xticks(valores, labels)
np_pops = np.array(populacoes['valores'])
np_pops = np_pops * 2 
plt.scatter(percapitas['valores'], expectativas['valores'], s=np_pops, c=cores['valores'], alpha=.8)
```

13. Coloque o nome dos paíse "Índia"e "China" próximo aos dois círulos maiores do gráfico scatter. Além disto, ative o grid do gráfico.

```python
import numpy as np
plt.figure(figsize=(20,10))
titulo="Desenvolvimento Humano em 2017"
eixox="Expectiva de vida em anos"
eixoy="Renda Percapita em US$"
plt.xscale('log')
plt.xlabel(eixox)
plt.ylabel(eixoy)
plt.title(titulo)
valores    = [1000, 10000, 100000]
labels     = ["1k", "10k", "100k"]
plt.xticks(valores, labels)
np_pops = np.array(populacoes['valores'])
np_pops = np_pops * 2 
plt.text(1950, 66, 'Índia')
plt.text(4700, 76, 'China')
plt.grid(True)
plt.scatter(percapitas['valores'], expectativas['valores'], s=np_pops, c=cores['valores'], alpha=.8)
```

17. Mostre em um gráfico de pizza a porecentagem de sobreviventes por sexo armazenados do dataframe **df_titanic**.

```python
somas = df_titanic['Survived'].groupby(df_titanic_sem_na['Sex']).sum()
plt.pie(somas, labels=somas.index, autopct='%.2f')
```

18. Mostre em um gráfico de pizza a porecentagem de não sobreviventes por sexo armazenados do dataframe **df_titanic**.

```python
df_nao_sobreviventes = df_titanic[df_titanic['Survived'] == 0]
somas = df_nao_sobreviventes['Survived'].groupby(df_titanic_sem_na['Sex']).count()
plt.pie(somas, labels=somas.index, autopct='%.2f')
```

19. Mostre a quantidade de não sobreviventes por classe econômica (Pclass).


```python
df_nao_sobreviventes = df_titanic[df_titanic['Survived'] == 0]
somas = df_nao_sobreviventes['Survived'].groupby(df_titanic_sem_na['Pclass']).count()
plt.pie(somas, labels=somas.index, autopct='%.2f')
```
