4. Importe o arquivo **seaslugs.txt** delimitado por tab(delimiter='\t') e armazene na variável **lugs**. Esse contém a porcentagem de larvas de lesma marinha que se metamorfoseou em um dado período de tempo. Examine as 10 primeiras linhas do arquivo com o comando **head** do linux.

```python
arquivo = 'datasets/seaslugs.txt'
!head -n 10 $arquivo
```

```python
lugs = np.loadtxt(arquivo, delimiter='\t', dtype=float, skiprows=1)
```

```python
lugs
```

5. Plote um gráfico scatter "Tempo (em min)" versus "Porcentagem de larvas".

```python
plt.figure(figsize=(20,10))
plt.scatter(lugs[:, 0], lugs[:, 1])
plt.xlabel('Tem (min.)')
plt.ylabel('Porcentagem de larvas')
```
