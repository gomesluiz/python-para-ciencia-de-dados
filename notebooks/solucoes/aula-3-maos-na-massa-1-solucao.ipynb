# Importação de arquivos textos

1. Importe o arquivo **datasets/moby_dick.txt** e armazene seu conteúdo na variável **texto**. Não utilize a cláusula **with**.

```python
nome = 'datasets/moby_dick.txt'
arquivo = open(nome, mode='r')
texto = arquivo.read()
arquivo.close()
```

2. Exiba o texto importado.

```python
print(texto)
```

3. Repita a importação anterior utilizando a cláusula **with**.

```python
with open(nome, mode='r') as arquivo:
    texto = arquivo.read()
```

4. Exiba o texto importado novamente.

```python
print(texto)
```

5. Importe e exiba apenas as cinco primeiras linhas do arquivo.

```python
with open(nome, mode='r') as arquivo:
    print(arquivo.readline())
    print(arquivo.readline())
    print(arquivo.readline())
    print(arquivo.readline())
    print(arquivo.readline())
```

6. (Desafio) Escreva um programa em Python que apresente a frequências das palavras no arquivo **moby_dick.txt**. Dica: pesquise os comandos **sub** e **split**.

```python
import re
# remove a pontuação do texto.
palavras = re.sub(r'[^\w\s]','',texto)

# quebra o texto em palavras.
palavras = palavras.lower().split()

# conta a frequência da palavras.
frequencias = {}
for palavra in palavras:
    if palavra in frequencias:
        frequencias[palavra] += 1 
    else:
        frequencias[palavra] = 1

# exibe a frequência das palavras.        
for palavra, frequencia in frequencias.items():
    print('{} :{}\n'.format(palavra, frequencia))
        
    
```
